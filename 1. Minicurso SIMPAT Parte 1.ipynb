# **Mas...o que é o Google Colab?**

Colaboratory, ou “Colab” de forma abreviada, é um produto da Google Research. Colab permite a qualquer um escrever e executar qualquer código em python atraves do browser, e é especialmente adequado para _machine learning_, análise de dados e educação. Mais tecnicamente, Colab é um serviço hospedado que provê Jupyter notebooks que não requerem qualquer configuração para uso e provê acesso gratuito a recursos computacionais incluindo GPUs.  ([Colaboratory FAQ](https://research.google.com/colaboratory/faq.html))

Além de python, atualmente o Colab também permite a execução de código em R, como veremos mais adiante.

# **Vantagens e Desvantagens do Google Colab:** 

✔ <u>Vantagens:</u>
 - Não há a necessidade de fazer instalação de programas;
 - Os _notebooks_ criados são facilmente compartilhados com outras pessoas.
 - **Reprodutibilidade de análise**

❎ <u>Desvantagens:</u>
  - O ambiente de edição não é tão elaborado quanto o de uma IDE como o RStudio por exemplo;
  - Os pacotes que não estejam pré instalados precisarão ser instalados a cada sessão;
  - Os arquivos de dados só ficam disponíveis por um determinado período;

# **Operando o Google Colab**

Agora vamos explorar um pouco os recursos que o Colab oferece para facilitar nosso trabalho de análise de dados. Isso, análise de dados. O Colab não é um editor de texto, embora possua, como vimos, alguns recursos para isso.

## **Células**

Conforme vocês já devem ter pecebido, o notebook é uma sucessão de células de dois tipos: **células de texto** e **células de código**.

A inserção dessas céluas no notebook pode ser feita de 4 maneiras:

* a partir do menu "Inserir" selecionando "Célula de código" ou "Célula de texto"
* a partir da barra superior do Colab nos botões **`+ Código`** ou **`+ Texto`** e
* a partir dos botões que surgem nas bordas superior e inferior das células quando passamos o mouse pelas bordas das células

Para **entrar no modo de edição** de uma **célula texto**, basta selecioná-la e pressionar a tecla **Enter** ou dar dois cliques. E para **sair do modo de edição** basta pressionar a tecla **Esc** ou selecionar outra célula.

Para executar o código em uma célula de código, basta clicar no botão "play", ou, alternativamente, com a célula selecionada clicar "CTRL+Enter".

Um recurso muito interessante quando estamos desenvolvendo um notebook em colaboração com outros(as) colegas é a possibilidade de incluir comentários nas células. 

Muito útil durante o desenvolvimento de um notebook, onde estamos experimentando as coisas é usar a "célula de código de rascunho".

### **_Dados disponibilizados na web_**

É possível capturar conjuntos de dados que estejam disponíveis na internet. Exemplo: a página do Prof. Pedro a. Morettin (https://www.ime.usp.br/~pam/) disponibiliza diversos conjuntos de dados utilizados em seus livros.
